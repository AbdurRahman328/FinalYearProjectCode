```python
def getData(path):
    data = pd.read_csv(path)
    data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
    data['Date'] = data['Date'].dt.strftime('%Y-%m-%d')
    data.set_index("Date",inplace=True)
    return data
```

```python
from skimage.restoration import (denoise_wavelet, estimate_sigma)

def denoiseData(data):
    data['Close']= denoise_wavelet(data.iloc[:, 0], wavelet='haar',
                                      #method='BayesShrink',
                                      method='VisuShrink',
                                      mode='soft', rescale_sigma = True)
    print(data[:15])
    return data
```

```python
def DatasetCreation(dataset, time_step = 1):  ##defining a function that gives a dataset and a time step, which then returns the input and output data
   DataX, DataY = [], []
   for i in range(len(dataset)- time_step):
         a = dataset[i:(i+ time_step), ]
         DataX.append(a)
         DataY.append(dataset[i + time_step, 0]) #ydata consists close price
   return np.array(DataX), np.array(DataY)


def data_split(data, split = 0.2):
  #======= creating training and test data===#
  l1   = int(len(data) * (1- split))
  l2    = len(data) - l1
  data1  = data.iloc[0:l1,:]
  data2   = data.iloc[l1:len(data),:]
  return data1, data2

def min_max_transform(data, feature_range=(0, 1)):
   scaler = MinMaxScaler(feature_range)
   return scaler.fit_transform(data)

def min_max_inverse_transform(data_scaled, min_original, max_original):
    return min_original +  data_scaled*(max_original - min_original)
```

```python
import tensorflow as tf

def load_model(model_path):
    try:
        model = tf.keras.models.load_model(model_path)
        print("Model loaded successfully.")
        return model
    except Exception as e:
        print("Error loading model:", str(e))
        return None
    
saved_model_path = "C:\\Users\\abdur\\Optimized-SingleLayer-LSTM-models\\optimized_single_layer_model_11.h5"
model = load_model(saved_model_path)
model.summary()
```

```python
from tkinter import ttk

def display_table(actual, predicted):
    # Create a frame within the main window
    table_frame = tk.Frame(root)
    table_frame.pack(pady=20)

    # Create a Treeview widget
    table = ttk.Treeview(table_frame)
    table["columns"] = ("actual", "predicted")
    table.heading("actual", text="Actual Price", anchor=tk.CENTER)
    table.heading("predicted", text="Predicted Price", anchor=tk.CENTER)

    # Set column alignment and enable grid lines
    table.column("actual", anchor=tk.CENTER)
    table.column("predicted", anchor=tk.CENTER)
    table["show"] = "headings"

    # Insert data into the table
    for i in range(len(actual)):
        table.insert("", i, text=str(i), values=(round(actual[i],4), predicted[i]))

    # Center-align cell values
    for col in table["columns"]:
        table.heading(col, anchor=tk.CENTER)

    table.pack(fill=tk.BOTH, expand=True)
```

```python
def perform_action(path):
    data = getData(path)
    data = denoiseData(data)

    # Scaling data
    min_test, max_test = data["Close"].min(), data["Close"].max()
    test_data_scaled = min_max_transform(data)

    # Creating input and output datasets for LSTM
    X_test, y_test = DatasetCreation(test_data_scaled, 5)
    print(X_test[:6])
    print("========================================================")
    print(y_test[:6])
    print("========================================================")
    num_features = test_data_scaled.shape[1]
    print("========================================================")
    

    # Inverting scaled target data to original scale
    y_test_original = min_max_inverse_transform(y_test, min_test, max_test)
    print(y_test_original[:6])
    
    ypred = min_max_inverse_transform(model.predict(X_test).ravel(), min_test, max_test)
    ytest = min_max_inverse_transform(y_test, min_test, max_test)
    
    mse =  np.mean((ytest - ypred)**2)
    rmse = np.sqrt(mse)
        
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")   
    display_table(ytest, ypred)
    
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.plot(y_test_original, label='Actual', color='blue')
    ax.plot(ypred, label='Predicted', color='orange')
    ax.set_ylabel('Price')
    ax.set_title('Actual vs Predicted Prices')
    ax.legend()

    # Embed the plot in the Tkinter window
    canvas = FigureCanvasTkAgg(fig, master=root)
    canvas.draw()
    canvas.get_tk_widget().pack()
```

```python
def upload_file():
    global file_path
    file_path = filedialog.askopenfilename()
    if file_path:
        print("File selected:", file_path)
        file_label.config(text="File Selected: " + file_path, fg="black")  # Update label with file path
```

```python
import tkinter as tk
from tkinter import filedialog
from tkinter import ttk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg


# Create the main window
root = tk.Tk()
root.title("Stock Market Prediction")
root.configure(bg="white")

# Create a button to trigger file upload
upload_button = tk.Button(root, text="Upload File", command=upload_file, bg="#007bff", fg="white", relief=tk.FLAT, borderwidth=0, padx=10, pady=5, font=("Helvetica", 12), cursor="hand2")
upload_button.pack(pady=20)

# Create a label to display the selected file path
file_label = tk.Label(root, text="", bg="white", fg="#007bff", font=("Helvetica", 10))
file_label.pack()

# Create a button to show results
show_results_button = tk.Button(root, text="Show Results", command=lambda: perform_action(file_path), bg="#007bff", fg="white", relief=tk.FLAT, borderwidth=0, padx=10, pady=5, font=("Helvetica", 12), cursor="hand2")
show_results_button.pack(pady=20)

# Add some padding
root.geometry("1080x1090")
root.resizable(True, True)

# Run the Tkinter event loop
root.mainloop()
```
